```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset 
data = pd.read_csv('C:\\Users\\Tanya Singh\\Desktop\\Data_science\\Task_1\\task1.csv', skiprows=4)


# View first few rows
print(data.head())

# Example: Bar chart of population for selected countries (latest year)
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil']
latest_year = '2023'  # Change this based on the data columns

subset = data[data['Country Name'].isin(selected_countries)][['Country Name', latest_year]]

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='Country Name', y=latest_year, data=subset)
plt.title(f"Population Distribution in {latest_year}")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
# Drop NaN values
pop_data = data[latest_year].dropna()

# Plot Histogram
plt.figure(figsize=(10,6))
sns.histplot(pop_data, bins=30, kde=True)
plt.title(f"Histogram of Population Across Countries ({latest_year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```
